```python
import os
from PIL import Image

def convert_images_to_grayscale(input_folder, output_folder):
    # Buat folder output jika belum ada
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    
    # Iterasi melalui semua file dalam folder input
    for filename in os.listdir(input_folder):
        print(filename)
        if filename.endswith(".jpg") or filename.endswith(".png"):
            # Path lengkap ke file gambar
            input_path = os.path.join(input_folder, filename)
            output_path = os.path.join(output_folder, filename)

            # Buka gambar dan ubah menjadi grayscale
            with Image.open(input_path) as img:
                grayscale_img = img.convert("L")
                grayscale_img.save(output_path)
            print(f"Converted {filename} to grayscale.")

# Contoh penggunaan
input_folder = "dataset_mentah/image"
output_folder = "dataset_mentah/image"
convert_images_to_grayscale(input_folder, output_folder)
```
